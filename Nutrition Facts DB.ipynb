```python
import pandas as pd
```

```python
nutritionFactsDF=pd.read_csv("nutrients_csvfile.csv")
nutritionFactsDF
```

```python
nutritionFactsDF.info()
```

```python
nutritionFactsDF.rename (columns={
    'Sat.Fat' : 'Saturated Fat'
    
}, inplace=True)
```

```python
nutritionFactsDF.head()
```

```python
nutritionFactsDF.rename (index={
    '(cornstarch)': 'Cornstarch'
    'or whipping' : 'Whipping Cream'
    '(1/2 cup ice cream)': 'Half a cup of ice cream'
}, inplace=True)
```

```python
nutritionFactsDF.head()
```

```python
from pprint import pprint
```

```python
NutritionFactsDict = nutritionFactsDF.to_dict('records')
NutritionFactsDict
```

```python
from pymongo import MongoClient
```

```python
mongo = MongoClient(port=27017)
```

```python
db = mongo.nutritionFactsDB
nutritionFactsData = db.nutritionFactsData
```

```python
nutritionFactsData.insert_many(NutritionFactsDict)
```

```python
results = list(nutritionFactsData.find())
```

```python
for order in results:
    pprint(order)
```
