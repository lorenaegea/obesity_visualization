```python
import pandas as pd
```

# Food with the most/least of each macro, based on category

```python
nutritionFactsDF=pd.read_csv("nutrients_csvfile.csv")
nutritionFactsDF
```

```python
nutritionFactsDF[['Protein','Fat', 'Sat.Fat', 'Fiber', 'Carbs']] = nutritionFactsDF[['Protein','Fat', 'Sat.Fat', 'Fiber', 'Carbs']].replace({'t': 0})
nutritionFactsDF
```

```python
print(nutritionFactsDF.columns)
```

```python
# Convert Calories column data type
nutritionFactsDF['Calories'] = pd.to_numeric(nutritionFactsDF['Calories'], errors='coerce')

# Group by category and find the index of the food with the maximum protein content
caloriesIndex = nutritionFactsDF.groupby('Category')['Calories'].idxmax()

# Use the index to select the corresponding rows and create a new dataframe
topCaloriesDF = nutritionFactsDF.loc[caloriesIndex, ['Food', 'Category', 'Calories']]

# Reset the index of the new dataframe
topCaloriesDF = topCaloriesDF.reset_index(drop=True)

topCaloriesDF
```

```python
# Convert Protein column data type
nutritionFactsDF['Protein'] = pd.to_numeric(nutritionFactsDF['Protein'], errors='coerce')

# Group by category and find the index of the food with the maximum protein content
proteinIndex = nutritionFactsDF.groupby('Category')['Protein'].idxmax()

# Use the index to select the corresponding rows and create a new dataframe
topProteinDF = nutritionFactsDF.loc[proteinIndex, ['Food', 'Category', 'Protein']]

# Reset the index of the new dataframe
topProteinDF = topProteinDF.reset_index(drop=True)

topProteinDF
```

```python
# Convert Fat column data type
nutritionFactsDF['Fat'] = pd.to_numeric(nutritionFactsDF['Fat'], errors='coerce')

# Group by category and find the index of the food with the maximum protein content
fatIndex = nutritionFactsDF.groupby('Category')['Fat'].idxmax()

# Use the index to select the corresponding rows and create a new dataframe
topFatDF = nutritionFactsDF.loc[fatIndex, ['Food', 'Category', 'Fat']]

# Reset the index of the new dataframe
topFatDF = topFatDF.reset_index(drop=True)

topFatDF
```

```python
# Convert Sat.Fat column data type
nutritionFactsDF['Sat.Fat'] = pd.to_numeric(nutritionFactsDF['Sat.Fat'], errors='coerce')

# Group by category and find the index of the food with the maximum protein content
satFatIndex = nutritionFactsDF.groupby('Category')['Sat.Fat'].idxmax()

# Use the index to select the corresponding rows and create a new dataframe
topSatFatDF = nutritionFactsDF.loc[satFatIndex, ['Food', 'Category', 'Sat.Fat']]

# Reset the index of the new dataframe
topSatFatDF = topSatFatDF.reset_index(drop=True)

topSatFatDF
```

```python
# Convert Fiber column data type
nutritionFactsDF['Fiber'] = pd.to_numeric(nutritionFactsDF['Fiber'], errors='coerce')

# Group by category and find the index of the food with the maximum protein content
fiberIndex = nutritionFactsDF.groupby('Category')['Fiber'].idxmax()

# Use the index to select the corresponding rows and create a new dataframe
topFiberDF = nutritionFactsDF.loc[fiberIndex, ['Food', 'Category', 'Fiber']]

# Reset the index of the new dataframe
topFiberDF = topFiberDF.reset_index(drop=True)

topFiberDF
```

```python
# Convert Carbs column data type
nutritionFactsDF['Carbs'] = pd.to_numeric(nutritionFactsDF['Carbs'], errors='coerce')

# Group by category and find the index of the food with the maximum protein content
carbsIndex = nutritionFactsDF.groupby('Category')['Carbs'].idxmax()

# Use the index to select the corresponding rows and create a new dataframe
topCarbsDF = nutritionFactsDF.loc[carbsIndex, ['Food', 'Category', 'Carbs']]

# Reset the index of the new dataframe
topCarbsDF = topCarbsDF.reset_index(drop=True)

topCarbsDF
```

```python
nutritionFactsDF.rename (columns={
    'Sat.Fat' : 'Saturated Fat'
    
}, inplace=True)
```
