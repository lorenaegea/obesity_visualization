```python
import pandas as pd
```

```python
from pprint import pprint
```

```python
healthFactorsDF =pd.read_csv("Nutrition__Physical_Activity__and_Obesity_-_Behavioral_Risk_Factor_Surveillance_System.csv")
healthFactorsDF
```

```python
healthFactorsDF.info()
```

```python
healthFactorsDF = healthFactorsDF.rename(columns={
    'LocationDesc' : 'Location',
    'YearStart':'Year',
    'Data_Value': 'Question_Percentage'
})
```

```python
healthFactorsDF2 = healthFactorsDF[["Year", "Location", "Question", "Question_Percentage", "Sample_Size", "Age(years)" ]]
healthFactorsDF2
```

```python
healthFactorsDF2 = healthFactorsDF2.loc[healthFactorsDF2['Question']=='Percent of adults who achieve at least 150 minutes a week of moderate-intensity aerobic physical activity or 75 minutes a week of vigorous-intensity aerobic physical activity and engage in muscle-strengthening activities on 2 or more days a week']
healthFactorsDF2
```

```python
mean = healthFactorsDF2.groupby(['Year', 'Age(years)'])['Question_Percentage'].mean()
mean = pd.DataFrame(mean)
mean
```

```python
mean = pd.merge(mean, healthFactorsDF2, on=['Year', 'Age(years)'], how='left')
mean = mean[mean['Location'] == 'Alabama']
mean
```

```python
mean = mean[['Year', 'Age(years)', 'Question_Percentage_x', 'Question', 'Sample_Size']]
mean = mean.rename(columns={'Question_Percentage_x':'Question_Percentage'})
mean = mean.reset_index(drop=True)
mean
```
